## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#Merging datasets
df1['Date'] = pd.to_datetime(df1['ActivityDate']).dt.date
df2['Date'] = pd.to_datetime(df2['ActivityHour']).dt.date
df3['Date'] = pd.to_datetime(df3['SleepDay']).dt.date
df4['Date'] = pd.to_datetime(df4['Date']).dt.date
````

````python
# Merging Daily_activity + Sleep
merged_df1=pd.merge(df1,df3,on=['Id','Date'],how='left')
````

````python
# Merging Weight Data
merged_df1= pd.merge(merged_df1,df4,on=['Id','Date'],how='left')
````

````python
#Merging hourlyStep data
hourly_daily = df2.groupby(['Id','Date'])['StepTotal'].sum().reset_index()
merged_df1 = merged_df1.merge(hourly_daily, on=['Id','Date'], how='left')
````

#### Chart - 16 merged_df1

````python
df_active = merged_df1.dropna(subset=['BMI', 'VeryActiveMinutes', 'FairlyActiveMinutes']).copy()
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Healthy'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'
df_active['BMICategory'] = df_active['BMI'].apply(categorize_bmi)
df_active['TotalActiveMinutes'] = df_active['VeryActiveMinutes'] + df_active['FairlyActiveMinutes']
plt.figure(figsize=(12, 6))
category_order = ['Underweight', 'Healthy', 'Overweight', 'Obese']
sns.boxplot(data=df_active, x='BMICategory', y='TotalActiveMinutes',order=category_order)
plt.title('Distribution of Active Minutes by BMI Category', fontsize=15)
plt.xlabel('BMI Category', fontsize=12)
plt.ylabel('Active Minutes (Very + Fairly)', fontsize=12)
plt.show()
````

#### Chart - 17 merged_df1

````python
cols_to_drop = ['Id', 'LogId', 'WeightPounds']
numeric_df = merged_df1.select_dtypes(include=['number']).drop(columns=cols_to_drop, errors='ignore')
corr_matrix = numeric_df.corr()
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='RdBu_r', center=0, linewidths=0.5)
plt.title('Correlation Heatmap: Activity, Sleep, and Weight Metrics', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.show()
````

#### Chart - 18 merged_df1

````python
merged_df1['ActivityLevel'] = pd.cut(merged_df1['TotalSteps'], 
                                     bins=[0, 5000, 10000, 20000], 
                                     labels=['Low', 'Medium', 'High'])
print(merged_df1['ActivityLevel'].value_counts())
````

````python
activity_comp = merged_df1.groupby('ActivityLevel', observed=False)[['Calories', 'TotalMinutesAsleep']].mean().reset_index()
fig, a1 = plt.subplots(figsize=(10, 6))
# Primary Axis: Calories (Bar Chart)
sns.barplot(data=activity_comp, x='ActivityLevel', y='Calories', ax=a1, palette='viridis', alpha=0.7)
a1.set_ylabel('Average Calories Burned', fontsize=12, color='darkgreen')
a1.set_xlabel('Activity Level (Steps)', fontsize=12)
# Secondary Axis: Sleep (Line Chart)
a2 = a1.twinx()
sns.lineplot(data=activity_comp, x='ActivityLevel', y='TotalMinutesAsleep', ax=a2, marker='o', color='red', linewidth=3, markersize=10)
a2.set_ylabel('Average Minutes Asleep', fontsize=12, color='red')
plt.title('Activity Level Impact: Calories vs. Sleep Duration', fontsize=16)
plt.show()
````
